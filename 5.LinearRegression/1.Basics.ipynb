```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Creating a function to work on
```

```python
X=np.arange(10)
Y=(X-5)**2
print(X,Y)
```

### Visualisation

```python
plt.style.use("seaborn")
plt.ylabel('F(X)')
plt.xlabel("X")
plt.plot(X,Y)
plt.show()
```

```python
x=1#any random no
lr=0.1#step size
error=[]
plt.plot(X,Y)
#we'll take 50 steps downhill
for i in range(50):
    grad=2*(x-5)
    x=x-lr*grad
    y=(x-5)**2
    error.append(y)
    print(x)
    plt.scatter(x,y)
#we observe value to be close to 5 which is the actual answer    
```

```python
plt.plot(error)#decreases untill 20 then grad becomes 0
```
