## Surface Plots
It is used for:-
- Visualise Loss Functions in Machine Learning & Deep Learning
- Visualise State or State Value Functions in Reinforcement Learning

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.array([1,2,3])
b=np.array([4,5,6,7,8])
a,b=np.meshgrid(a,b)
print(a)#copy elements of a row wise n times , n=no. of elements in b
print(b)#copy elements of b column wise n times , n=no. of elements in a
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')#x,y,z axis
plt.show()#red- high value, blue-low value region
```

```python
a = np.arange(-1,1,0.02)
b = a 
a,b = np.meshgrid(a,b)
```

```python
fig = plt.figure()#at 0 - it will be minima z=a^2+b^2
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a,b,a**2+b**2,cmap='rainbow')
plt.title("Contour Plot")
plt.show()#same value of z forms contour plots
```
