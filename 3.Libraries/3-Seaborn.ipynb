```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips) #which gender generates a higher bill?,generates mean
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)#plots standard deviation
```

```python
sns.countplot(x = 'sex', data= tips) #bar graph- how many males and females are there? 
```

```python
sns.countplot(x = 'time', data = tips )
```

```python
sns.boxplot(x="day" , y='total_bill' , data= tips, hue='sex')
```

```python
sns.boxplot(x="day" , y='total_bill' , data= tips) #the horizontal lines depict the highest and lowest bill the box denotes where majority of the data lies, and the dots represent outliers
```

```python
sns.boxplot(x="day" , y='total_bill' , data= tips,hue='smoker') 
```

```python
sns.violinplot(x='day', y = 'total_bill', data = tips, hue = 'sex', split = True)
```

```python
sns.distplot(tips['total_bill'], kde = False, bins = 30)#show distibution of one variable
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x= 'total_bill' , y='tip', data = tips, kind='hex' )
```

```python
sns.pairplot(tips, hue="sex")#all combinations
```
